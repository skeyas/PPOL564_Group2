```python
import os
import re
import requests
import pprint
import pandas as pd
```

```python
all_files = os.listdir(os.getcwd())    
csv_files = list(filter(lambda f: f.endswith('.csv'), all_files))

file_dict = {re.findall(r"\d+", f)[0]: f for f in csv_files}
```

```python
pd.set_option('display.max_columns', None)


data_frames = [pd.read_csv(file_dict[file]) for file in file_dict.keys()]
tmp = pd.concat(data_frames)
tmp["20. INDUSTRY SECTOR"].unique()
tmp.head()
```
